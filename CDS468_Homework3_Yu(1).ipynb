```python
# Homework 3a / problem 1: Use the image of your choosing, show your image as a grayscale: 7 bit, 5 bit, and 3 bit.

%matplotlib inline

import matplotlib.pyplot as plt
import imageio
import numpy as np
import urllib

url = 'https://cdn.mos.cms.futurecdn.net/qTmvroik2NtHVGsfCHyyxF-1200-80.jpg'
mgdata = imageio.imread(url, as_gray=True).astype(np.uint8)

#showing original image as greyscale / commented out to only show 7 bit, 5 bit, and 3 bit images. 

#plt.title('Gray Scale image')
#plt.imshow(mgdata, cmap='gray')
#plt.show()

#showing original image as 7-bit 

img_7bit = mgdata & 0x80 # 128 as hex is 0x80
plt.title('7 bit grey image')
plt.imshow(img_7bit, cmap = 'gray')
plt.show()

#showing original image as 5-bit 

img_5bit = mgdata & 0x20 #32 as hex is 0x20
plt.title('5 bit grey image')
plt.imshow(img_5bit, cmap = 'gray')
plt.show()

#showing original image as 3 bit

img_3bit = mgdata & 0x8 #8 as hex is 8
plt.title('3 bit grey image')
plt.imshow(img_3bit, cmap = 'gray')
plt.show()
```
